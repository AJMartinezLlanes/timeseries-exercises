```python
import pandas as pd

# The requests library simplifies the process of making http requests
import requests
```

1. Using the code from the lesson as a guide and the REST API from https://api.data.codeup.com/api/v1/items as we did in the lesson, create a dataframe named items that has all of the data for items.

```python
base_url = 'https://api.data.codeup.com'
endpoint = '/api/v1/items'
items = []

while endpoint != None:
    url = base_url + endpoint
    response = requests.get(url)
    data = response.json()
    items.extend(data['payload']['items'])
    current_page = data['payload']['page']
    max_page = data['payload']['max_page']
    endpoint = data['payload']['next_page']
    print(f'current_page: {current_page}')
    print(f'max_page: {max_page}')
    print(f'next_page: {endpoint}')
items = pd.DataFrame(items)
```

```python
items.shape
```

2. Do the same thing, but for stores (https://api.data.codeup.com/api/v1/stores)

```python
response = requests.get('https://api.data.codeup.com/api/v1/stores')
data = response.json()
stores = pd.DataFrame(data['payload']['stores'])
stores.head()
```

```python
stores.shape
```

3. Extract the data for sales (https://api.data.codeup.com/api/v1/sales). There are a lot of pages of data here, so your code will need to be a little more complex. Your code should continue fetching data from the next page until all of the data is extracted.

```python
base_url = 'https://api.data.codeup.com'
endpoint = '/api/v1/sales'
sales = []

while endpoint != None:
    url = base_url + endpoint
    response = requests.get(url)
    data = response.json()
    sales.extend(data['payload']['sales'])
    current_page = data['payload']['page']
    max_page = data['payload']['max_page']
    endpoint = data['payload']['next_page']
    print(f'current_page: {current_page}')
    print(f'max_page: {max_page}')
    print(f'next_page: {endpoint}')
sales = pd.DataFrame(sales)
```

```python
sales.shape
```

4. Save the data in your files to local csv files so that it will be faster to access in the future.


```python
items.to_csv('items.csv',index=False)
stores.to_csv('stores.csv',index=False)
sales.to_csv('sales.csv',index=False)
```

5. Combine the data from your three separate dataframes into one large dataframe.

```python
df = sales.merge(items, how='left', left_on='item', right_on='item_id').drop(columns=['item'])
df.head()
```

```python
df = df.merge(stores, how='left', left_on='store', right_on='store_id').drop(columns=['store'])
df.head()
```
